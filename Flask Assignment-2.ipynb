Q1. Explain GET and POST methods

GET and POST are two common HTTP methods:

GET: This method is used to request data from a specified resource. The data sent to the server is appended to the URL. Itâ€™s important to note that GET requests can be cached, remain in the browser history, and can be bookmarked. However, they have length restrictions and should never be used when dealing with sensitive data.

POST: This method is used to send data to the server to create or update a resource. The data sent to the server with POST is stored in the request body of the HTTP request. Unlike GET requests, POST requests are never cached, do not remain in the browser history, cannot be bookmarked, and have no restrictions on data length.



Q2. Why is request used in Flask?

In Flask, the request object contains all the data sent from the client to the server. This data can be recovered using the GET/POST methods. The request object is crucial as it allows you to access incoming request data such as form data, query parameters, JSON payloads, and file uploads.



Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect the user to a different endpoint with a specified status code. This is particularly useful when a certain action is performed by the user, or when the requested URL is no longer available.



Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are files that contain static data as well as placeholders for dynamic data. They allow you to generate HTML dynamically and present it to the user. Flask uses the Jinja template engine to render templates.

The render_template() function in Flask is used to generate output from a template file based on the Jinja engine that is populated with data from your application. It allows you to send variables from your Python script to your HTML file, where you can use them to dynamically populate the HTML.
